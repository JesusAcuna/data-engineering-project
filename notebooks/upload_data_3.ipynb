```python
def make_df(year,category,month_ranges,dep):
    df=pd.DataFrame()
    
    if year=="2022":
        names_en_1=['Department',
        'ITE_DC:Total','ITE_DC:Cases','ITE_DC:Rate',
        'ITE_RDC:Total','ITE_RDC:Cases','ITE_RDC:Rate',
        'IPE_DG:Total','IPE_DG:Cases','IPE_DG:Rate',
        'IPT_DA:Total','IPT_DA:Cases','IPT_DA:Rate',
        'IPT:Total','IPT_RDA:Cases','IPT_RDA:Rate','IPT_SP:Cases','IPT_SP:Rate','IPT_O:Cases','IPT_O:Rate']
        
        names_en_2=['Department','Province','District','Ubigeous',
        'ITE_DC:Total','ITE_DC:Cases','ITE_DC:Rate',
        'ITE_RDC:Total','ITE_RDC:Cases','ITE_RDC:Rate',
        'IPE_DG:Total','IPE_DG:Cases','IPE_DG:Rate',
        'IPT_DA:Total','IPT_DA:Cases','IPT_DA:Rate',
        'IPT_RDA:Total','IPT_RDA:Cases','IPT_RDA:Rate','IPT_SP:Cases','IPT_SP:Rate','IPT_O:Cases','IPT_O:Rate']
        
        names_an_1=['Department','Total',
       'AT:Cases','AT:Rate',
       'AL:Cases','AL:Rate',
       'AM:Cases','AM:Rate',
       'AS:Cases','AS:Rate']
        
        names_an_2=['Department','Province','District','Ubigeous','Total',
       'AT:Cases','AT:Rate',
       'AL:Cases','AL:Rate',
       'AM:Cases','AM:Rate',
       'AS:Cases','AS:Rate']
        
        names_imc_clap_1=['Department',
        'IMC_PG:Total','IMC_PG_BP:Cases','IMC_PG_BP:Rate','IMC_PG_SP:Cases','IMC_PG_SP:Rate','IMC_PG_O:Cases','IMC_PG_O:Rate',
        'CLAP:Total','CLAP_DP:Cases','CLAP_DP:Rate','CLAP_SP:Cases','CLAP_SP:Rate']
        
        names_imc_clap_2=['Department','Province','District','Ubigeous',
        'IMC_PG:Total','IMC_PG_BP:Cases','IMC_PG_BP:Rate','IMC_PG_SP:Cases','IMC_PG_SP:Rate','IMC_PG_O:Cases','IMC_PG_O:Rate',
        'CLAP:Total','CLAP_DP:Cases','CLAP_DP:Rate','CLAP_SP:Cases','CLAP_SP:Rate']

        
        
        if category=="children" :
            if (("EN" in dep)and("DEP" in dep)):    
                print(f"{year}_{category}_{dep}")        
                skiprows=7
                usecols="B:U"
                names=names_en_1
                nrows=25
                skipfooter=0
            elif (("EN" in dep)and("DIRESA" in dep)):    
                print(f"{year}_{category}_{dep}")        
                skiprows=7
                usecols="B:U"
                names=names_en_1
                nrows=29
                skipfooter=0                
            elif ("EN" in dep) and ("DISTRITO" in dep):
                print(f"{year}_{category}_{dep}")        
                skiprows=7
                usecols="B:X"
                names=names_en_2
                nrows=None
                skipfooter=7                
            elif (("Anemia" in dep) and ("DEP" in dep)) :
                print(f"{year}_{category}_{dep}")        
                usecols='B:K'
                skiprows=6
                names=names_an_1
                nrows=25
                skipfooter=0
            elif (("Anemia" in dep) and ("DIRESA" in dep)) :
                print(f"{year}_{category}_{dep}")        
                usecols='B:K'
                skiprows=6
                names=names_an_1
                nrows=29
                skipfooter=0
            elif (("Anemia" in dep) and ("DISTRITO" in dep)) :
                print(f"{year}_{category}_{dep}")        
                usecols='B:N'
                skiprows=6
                names=names_an_2
                nrows=None
                skipfooter=5
                
            df=pd.read_excel(f"./data/{year}/{category}/{category}_{year}_{month_ranges}.xlsx",
                             sheet_name=dep,
                             usecols=usecols,
                             skiprows=skiprows,
                             nrows=nrows,
                             skipfooter=skipfooter,
                             engine='openpyxl',
                             header=None,
                             names=names)
            
        elif category=="foreign_children":
            
            if (("EN" in dep)and("DEP" in dep)):    
                print(f"{year}_{category}_{dep}")        
                skiprows=7
                usecols="B:U"
                names=names_en_1
                nrows=25
                skipfooter=0
            elif (("EN" in dep)and("DIRESA" in dep)):    
                print(f"{year}_{category}_{dep}")        
                skiprows=7
                usecols="B:U"
                names=names_en_1
                nrows=29
                skipfooter=0
            elif ("EN" in dep) and ("DISTRITO" in dep):
                print(f"{year}_{category}_{dep}")        
                skiprows=7
                usecols="B:X"
                names=names_en_2
                nrows=None
                skipfooter=5
            elif (("Anemia" in dep) and ("DEP" in dep)) :
                print(f"{year}_{category}_{dep}")        
                usecols='B:K'
                skiprows=6
                names=names_an_1
                nrows=25
                skipfooter=0
            elif (("Anemia" in dep) and ("DIRESA" in dep)) :
                print(f"{year}_{category}_{dep}")        
                usecols='B:K'
                skiprows=6
                names=names_an_1
                nrows=29
                skipfooter=0
            elif (("Anemia" in dep) and ("DISTRITO" in dep)) :
                print(f"{year}_{category}_{dep}")        
                usecols='B:N'
                skiprows=6
                names=names_an_2
                nrows=None
                skipfooter=3
                
            df=pd.read_excel(f"./data/{year}/{category}/{category}_{year}_{month_ranges}.xlsx",
                             sheet_name=dep,
                             usecols=usecols,
                             skiprows=skiprows,
                             nrows=nrows,
                             skipfooter=skipfooter,
                             engine='openpyxl',
                             header=None,
                             names=names)
        elif category=="VRAEM_children":    
            
            if (("EN" in dep) and ("DISTRITO" in dep)):
                print(f"{year}_{category}_{dep}")        
                skiprows=7
                usecols="B:X"
                names=names_en_2
                nrows=None
                skipfooter=6
            elif (("Anemia" in dep) and ("DISTRITO" in dep)) :
                print(f"{year}_{category}_{dep}")        
                usecols='B:N'
                skiprows=6
                names=names_an_2
                nrows=None
                skipfooter=4
                
            df=pd.read_excel(f"./data/{year}/{category}/{category}_{year}_{month_ranges}.xlsx",
                             sheet_name=dep,
                             usecols=usecols,
                             skiprows=skiprows,
                             nrows=nrows,
                             skipfooter=skipfooter,
                             engine='openpyxl',
                             header=None,
                             names=names)        
        elif category=="pregnant_women":
            
            if (("IMC_PG-CLAP" in dep)and("DEP" in dep)):    
                print(f"{year}_{category}_{dep}")        
                skiprows=7
                usecols="B:N"
                names=names_imc_clap_1
                nrows=25
                skipfooter=0
                
            elif (("IMC_PG-CLAP" in dep)and("DIRESA" in dep)):    
                print(f"{year}_{category}_{dep}")        
                skiprows=7
                usecols="B:N"
                names=names_imc_clap_1
                nrows=34
                skipfooter=0  
                
            elif (("IMC_PG-CLAP" in dep) and ("DISTRITO" in dep)):
                print(f"{year}_{category}_{dep}")        
                skiprows=7
                usecols="B:Q"
                names=names_imc_clap_2
                nrows=None
                skipfooter=7
            
            elif (("Anemia" in dep) and ("DEP" in dep)) :
                print(f"{year}_{category}_{dep}")        
                usecols='B:K'
                skiprows=6
                names=names_an_1
                nrows=25
                skipfooter=0
            elif (("Anemia" in dep) and ("DIRESA" in dep)) :
                print(f"{year}_{category}_{dep}")        
                usecols='B:K'
                skiprows=6
                names=names_an_1
                nrows=34
                skipfooter=0                
            elif (("Anemia" in dep) and ("DISTRITO" in dep)) :
                print(f"{year}_{category}_{dep}")        
                usecols='B:N'
                skiprows=6
                names=names_an_2
                nrows=None
                skipfooter=6
                
            df=pd.read_excel(f"./data/{year}/{category}/{category}_{year}_{month_ranges}.xlsx",
                             sheet_name=dep,
                             usecols=usecols,
                             skiprows=skiprows,
                             nrows=nrows,
                             skipfooter=skipfooter,
                             engine='openpyxl',
                             header=None,
                             names=names)                    

    return df
```

```python
from openpyxl import load_workbook
year="2022"
category="children"
month_ranges="01_03"

try:
    wb = load_workbook(filename = f"./data/{year}/{category}/{category}_{year}_{month_ranges}.xlsx")

    sheets= wb.sheetnames[1:]

    for sheet in sheets:
        dep=sheet
        print(dep)
        df=make_df(year=year,category=category,month_ranges=month_ranges,dep=dep)
        df.replace(['*','(SD)'], 0 , inplace=True)
        df.fillna(method="ffill",inplace=True)
        display(df)
        
except FileNotFoundError:
    print("no such file or directory")
```
