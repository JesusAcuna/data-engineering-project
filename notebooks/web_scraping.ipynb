```python
import requests
import os
from collections import defaultdict
from bs4 import BeautifulSoup
```

```python
def convert_month_to_number(value_name):
    Months=['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Setiembre','Septiembre','Octubre','Noviembre','Diciembre']
    Months_map=['01','02','03','04','05','06','07','08','09','09','10','11','12']

    months_dict=dict(zip(Months, Months_map))

    names_dict_new=[]

    for month in Months:
        if month in value_name:
            names_dict_new.append(months_dict.get(month,"There's no value for that month in 'months_dict'"))    
    
    num_dict_new = ""
    for i in names_dict_new:
        num_dict_new=num_dict_new+i+"_"
        
    return num_dict_new[:-1]
```

```python
def web_scraping(URL,years):
    """
    URL: website URL 
    years: is a list of ranges of years
    
    """
    ###
    html_text=requests.get(URL).text
    soup=BeautifulSoup(html_text,'lxml')
    block=soup.find("main",class_="l-bl block block-main-content") 
    li_tags= block.find_all('li')

    first_filter = "Indicadores"

    year_list=years
    second_filter=[str(i)for i in year_list]

    names_dict=defaultdict(list)
    urls_dict=defaultdict(list)

    for li_tag in li_tags:
        li_text = li_tag.a.text.replace("\xa0", " ")
        if first_filter in li_text:
            if any((match := substring) in li_text for substring in second_filter):
                names_dict[match].append(li_text)
                li_url = li_tag.a['href']
                urls_dict[match].append(li_url)         
    ###
            
    if not os.path.exists("./data"):
        os.mkdir("./data")  

    for i in list(names_dict.keys()):
        if not os.path.exists(f"./data/{i}"):
            os.mkdir(f"./data/{i}")
        if not os.path.exists(f"./data/{i}/pregnant_women"):
            os.mkdir(f"./data/{i}/pregnant_women")
        if not os.path.exists(f"./data/{i}/children"):
            os.mkdir(f"./data/{i}/children") 

    years = names_dict.keys()


    for year in years:
        print(f"{year}:")

        value_name_list=names_dict.get(year,"there's no value_name for that year")
        value_url_list=urls_dict.get(year,"there's no value_url for that year")

        for i,value_name in enumerate(value_name_list):
            num_dict_new=convert_month_to_number(value_name)

            if "Gestantes" in value_name:
                if "web.ins.gob.pe" in value_url_list[i]:
                    URL=value_url_list[i]
                    print("pregnant_women if")
                    print(URL)
                    file_name=f"pregnant_women_{year}_{num_dict_new}"
                    response = requests.get(URL)
                    open(f"./data/{year}/pregnant_women/{file_name}.xlsx", "wb").write(response.content)

                    continue
                else:
                    URL="https://web.ins.gob.pe"+value_url_list[i]
                    print("pregnant_women else")
                    print(URL)
                    file_name=f"pregnant_women_{year}_{num_dict_new}"
                    response = requests.get(URL)
                    open(f"./data/{year}/pregnant_women/{file_name}.xlsx", "wb").write(response.content)

                    continue

            elif ("Niños Venezolanos" or "Niños Extranjeros") in value_name:
                if not os.path.exists(f"./data/{year}/foreign_children"):
                    os.mkdir(f"./data/{year}/foreign_children") 

                if "web.ins.gob.pe" in value_url_list[i]:
                    URL=value_url_list[i]
                    print("foreign_children if")
                    print(URL)
                    file_name=f"foreign_children_{year}_{num_dict_new}"
                    response = requests.get(URL)
                    open(f"./data/{year}/foreign_children/{file_name}.xlsx", "wb").write(response.content)

                    continue
                else:
                    URL="https://web.ins.gob.pe"+value_url_list[i]
                    print("foreing_children else")
                    print(URL)
                    file_name=f"foreign_children_{year}_{num_dict_new}"
                    response = requests.get(URL)
                    open(f"./data/{year}/foreign_children/{file_name}.xlsx", "wb").write(response.content)

                    continue
            elif "Niños VRAEM" in value_name:
                if not os.path.exists(f"./data/{year}/VRAEM_children"):
                    os.mkdir(f"./data/{year}/VRAEM_children") 

                if "web.ins.gob.pe" in value_url_list[i]:
                    URL=value_url_list[i]
                    print("VRAEM_children if")
                    print(URL)
                    file_name=f"VRAEM_children_{year}_{num_dict_new}"
                    response = requests.get(URL)
                    open(f"./data/{year}/VRAEM_children/{file_name}.xlsx", "wb").write(response.content)

                    continue
                else:
                    URL="https://web.ins.gob.pe"+value_url_list[i]
                    print("VRAEM_children else")
                    print(URL)
                    file_name=f"VRAEM_children_{year}_{num_dict_new}"
                    response = requests.get(URL)
                    open(f"./data/{year}/VRAEM_children/{file_name}.xlsx", "wb").write(response.content)

                    continue
            else:
                if "web.ins.gob.pe" in value_url_list[i]:
                    URL=value_url_list[i]
                    print("children if")
                    print(URL)
                    file_name=f"children_{year}_{num_dict_new}"
                    response = requests.get(URL)
                    open(f"./data/{year}/children/{file_name}.xlsx", "wb").write(response.content)

                    continue

                elif "gob.pe" in value_url_list[i]:
                    URL="https://cdn.www.gob.pe/uploads/document/file/3566498/1.Indic%20Ni%C3%B1os%20a%20Junio%202022%20-%20PERU.xlsx?v=1661961860"
                    print("children elif")
                    print(URL)
                    file_name=f"children_{year}_{num_dict_new}"
                    response = requests.get(URL)
                    open(f"./data/{year}/children/{file_name}.xlsx", "wb").write(response.content)

                    continue
                else:
                    URL="https://web.ins.gob.pe"+value_url_list[i]
                    print("children else")
                    print(URL)
                    file_name=f"children_{year}_{num_dict_new}"
                    response = requests.get(URL)
                    open(f"./data/{year}/children/{file_name}.xlsx", "wb").write(response.content)

                    continue

        print()
    print("End of program")
```
